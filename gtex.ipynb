```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
```

```python
muscle_tpm = pd.read_csv('gene_tpm_muscle_skeletal.gct', sep='\t', skiprows=2)
```

```python
muscle_tpm
```

```python
ensembleID = pd.read_csv('ensembleID', sep='\t')
```

```python
ensembleID
```

```python
ensembleID['#geneName'] = ensembleID['#geneName'].str.split('.').str[0]
```

```python
muscle_tpm['Name']=muscle_tpm['Name'].str.split('.').str[0]
```

```python
ensembleID = ensembleID.rename(columns={"#geneName": "Name"})
```

```python
muscle_tpm = pd.merge(muscle_tpm, ensembleID, on=['Name'], how='right')
```

```python
muscle_tpm
```

```python
numeric_cols = muscle_tpm.iloc[:, 2:].select_dtypes(include=[np.number])
muscle_tpm['Median_TPM'] = numeric_cols.median(axis=1)
thresholds = muscle_tpm['Median_TPM'].quantile([0.33, 0.67]).values
muscle_tpm['Groupe'] = muscle_tpm['Median_TPM'].apply(lambda x: 'Groupe 1' if x < thresholds[0] else ('Groupe 2' if x < thresholds[1] else 'Groupe 3'))

melted_TPM = muscle_tpm.melt(id_vars=['Name', 'Description', 'Groupe'], var_name='SampleID', value_name='TPM')
melted_TPM.sort_values(by=['Groupe', 'Description'], inplace=True)
```

```python
with PdfPages('muscle_GTEx_TPM.pdf') as pdf:
    plt.figure(figsize=(12, 50))
    ax = sns.stripplot(data=melted_TPM, y='Description', x='TPM', jitter=0.25, size=2, alpha=0.5, hue='Groupe')
    plt.xscale('log')

    ticks = [10**i for i in range(0, 6)]
    labels = [f"{10**i}" for i in range(0, 6)]
    plt.xticks(ticks, labels)

    plt.xlabel('TPM')
    plt.grid(True)
    ax.set_xlim(left=1)
    ax.set_ylabel('')
    ax.legend_.remove()
    
    pdf.savefig()
    plt.close()
```
